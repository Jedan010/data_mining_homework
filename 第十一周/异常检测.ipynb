```python
data = np.array([24.0, 28.9, 28.9, 29.0, 29.1, 29.1, 29.2, 29.2, 29.3, 29.4])
```

```python
plt.figure(dpi=100)
plt.scatter(np.arange(len(data)), data)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(data.reshape(-1,1))
```

```python
plt.figure(dpi=100)

plt.scatter(np.arange(len(data[kmeans.labels_==1])), data[kmeans.labels_==1], label='normal')
plt.scatter(np.arange(len(data[kmeans.labels_==0])), data[kmeans.labels_==0], label='outlier')
plt.legend()
```

```python
data = pd.read_excel('default of credit card clients.xls', header=1)
data.head()
```

```python
X = data.drop(data.columns[-1], axis=1).values
y = data[data.columns[-1]].values

y_ = y.copy()
y_[y==0] = -1
```

# Kmeans

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data.drop(data.columns[-1], axis=1))
y_kmean = kmeans.labels_
```

# LOF

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors=40)
y_pred = clf.fit_predict(X)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

# 孤立森林

```python
from sklearn.ensemble import IsolationForest
# fit the model
clf = IsolationForest(max_samples=100)
clf.fit(X)
y_pred_outliers = clf.predict(X)
```

```python
print(classification_report(y, y_kmean))
```

```python
print(classification_report(y_, y_pred))
```

```python
print(classification_report(y_, y_pred_outliers))
```
