## 数据类型转换

### 将有序文本转换为数值

```python
df_tranfer = df.copy()
for column, map_value in maping.items():
    df_tranfer[column] = df[column].map(map_value)
df_tranfer.head()
```

### 将无序文本转换为哑变量（one-hot）

```python
from sklearn.preprocessing import OneHotEncoder
for column in df_tranfer.select_dtypes(include=['object']).columns:
    df_tranfer[column] = np.unique(df_tranfer[column], return_inverse=True)[1]
    for i in range(pd.value_counts(df_tranfer[column]).count()):
        df_tranfer[column+'_'+str(i+1)] =  OneHotEncoder().fit_transform(df_tranfer[column].values.reshape((-1,1))).toarray()[:,i]
    df_tranfer.drop(column, axis=1, inplace=True)
df_tranfer.head()
```

```python
df_tranfer.info()
```

### 增加特征

```python
df_tranfer['saving_by_credit'] = df_tranfer['Savings_account'] / df_tranfer['Credit_amount']
df_tranfer['each_loan'] =   df_tranfer['Credit_amount'] / (df_tranfer['guarantors']+1)
df_tranfer.head()
```

## 特征筛选

### RFE筛选特征

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
X_val = df_tranfer.drop('label', axis=1)
y_val = df_tranfer['label']
rfe = RFE(estimator=LogisticRegression(), n_features_to_select=20).fit(X_val, y_val)
rfe.support_
```

查看被筛选掉的特征

```python
X_val.columns[rfe.support_ == False]
```

```python
X_val_filted = X_val.iloc[:, rfe.support_]
X_val_filted.head()
```

### 相关性筛选

```python
plt.figure(figsize=(12,12))
plt.title('各个特征的皮尔逊相关系数', y=1.05, size=15)
sns.heatmap(X_val_filted.corr(),linewidths=0.1,vmax=1.0, square=True, cmap=plt.cm.viridis, linecolor='white', annot=True)
```

```python
column_to_filt = ['Personal_status_and_sex_3', 'Personal_status_and_sex_2', 'Other_installment_plans_1', 'Other_installment_plans_3']
X_val_filted = X_val_filted.drop(labels=column_to_filt, axis=1)
```

```python
plt.figure(figsize=(12,12))
plt.title('各个特征的皮尔逊相关系数', y=1.05, size=15)
sns.heatmap(X_val_filted.corr(),linewidths=0.1,vmax=1.0, square=True, cmap=plt.cm.viridis, linecolor='white', annot=True)
```

### Embedded

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=10)#构建分类随机森林分类器
clf.fit(X_val_filted, y_val) #对自变量和因变量进行拟合
for i, score in enumerate(clf.feature_importances_):
    print(X_val_filted.columns[i], '\t\t',score)
    if score < 0.01:
        print()
        print('这个特征不要了:', X_val_filted.columns[i])
        print()
```

```python
## feature importances 可视化##
importances = clf.feature_importances_
feat_names = X_val_filted.columns
indices = np.argsort(importances)[::-1]
fig = plt.figure(figsize=(20,6))
plt.title("Feature importances by RandomTreeClassifier")
plt.bar(range(len(indices)), importances[indices], color='lightblue',  align="center")
plt.step(range(len(indices)), np.cumsum(importances[indices]), where='mid', label='Cumulative')
plt.xticks(range(len(indices)), feat_names[indices], rotation='vertical',fontsize=14)
plt.xlim([-1, len(indices)])
plt.show()
```

## 过采样

```python
pd.value_counts(y_val)
```

```python
# 处理不平衡数据
from imblearn.over_sampling import SMOTE # 导入SMOTE算法模块
sm = SMOTE(random_state=42)    # 处理过采样的方法
X_resample, y_resample = sm.fit_sample(X_val_filted, y_val)
```

```python
pd.value_counts(y_resample)
```

## 网格搜索和交叉验证

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resample, y_resample)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
k_range = list(range(1,20))
weight_options = ['uniform', 'distance']
kernel = ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']
algorithm = ['auto', 'ball_tree', 'kd_tree', 'brute']
param_knn = dict(n_neighbors=k_range, weights=weight_options, algorithm=algorithm)

RSCV_knn = RandomizedSearchCV(knn, param_knn, cv=10, scoring='accuracy',n_iter=10, random_state=5)
RSCV_knn.fit(X_train, y_train)
print("knn_best:%f" %RSCV_knn.best_score_)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
y_knn_pred = RSCV_knn.predict(X_test)
print('accuracy:', accuracy_score(y_test, y_knn_pred))
print(classification_report(y_test, y_knn_pred))
```

```python
from plot_comfusion_matrix import *
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_knn_pred)    
classs = ['good','bad']
plot_confusion_matrix(cm, classes=classs,normalize=True, title='Confusion matrix of best KNN')
```
